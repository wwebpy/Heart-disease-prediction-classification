id = id

age = Alter

sex = Geschlecht

dataset = herkunft datensatz

cp = Brustschmerz-Typ
     0 = asymptomatisch
     1 = typische Angina
     2 = atypische Angina
     3 = nicht-anginöser Schmerz

trestbps = Ruhe-Blutdruck (in mm Hg)

chol = Serum-Cholesterin (in mg/dl)

fbs = Nüchternblutzucker > 120 mg/dl
      0 = nein
      1 = ja

restecg = Ruhe-EKG-Ergebnis
          0 = normal
          1 = ST-T-Wellen-Anomalie
          2 = linksventrikuläre Hypertrophie

thalch (thalach) = Maximale erreichte Herzfrequenz

exang = Belastungsinduzierte Angina
        0 = nein
        1 = ja

oldpeak = ST-Senkung bei Belastung relativ zur Ruhe
          → Maß für belastungsbedingte Herzbelastung

slope = Steigung des ST-Segments bei Belastung
        0 = ansteigend
        1 = flach
        2 = abfallend

ca = Anzahl der großen Gefäße (0–3), die durch Fluoroskopie sichtbar sind

thal = Thalassämie-Test-Ergebnis
       1 = normal
       2 = fixer Defekt
       3 = reversibler Defekt

num = TARGET = Diagnose der Herzkrankheit
      0 = keine Herzkrankheit
      1–4 = Herzkrankheit vorhanden (unterschiedlicher Schweregrad)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('dark_background')
sns.set_palette("viridis")
```

```python
data = pd.read_csv('../data/raw/heart_disease_uci.csv', sep=',')
data.head(100)
```

```python
data.info()
```

```python
data = data.drop(columns=['id', 'dataset'])
data.info()
```

```python
data.head()
```

```python
# Alle Missing Values und NAN für alle spalten anzeigen
data.isnull().sum()
```

```python
# Fehlende Werte entfernen & ersetzen
data = data.drop(columns=['ca', 'thal'])

data['slope_missing'] = data['slope'].isnull().astype(int)
data['slope'] = data['slope'].fillna(data['slope'].mode()[0])
data['slope'].value_counts(dropna=False)
```

```python
data['slope'].isnull().sum()
```

```python
# Zielvariable 'num' zu 'target' formatieren
data['target'] = (data['num'] > 0).astype(int)
data = data.drop(columns=['num'])
data.head()
```

```python
data.info()
```

```python
# Dummy encoding
data = pd.get_dummies(data, columns=['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope'], drop_first=True)
data.head()
```

```python
data.to_csv('../data/processed/heart_disease_dummyEncoding.csv', index=False)
```
